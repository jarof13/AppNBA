# DuckDB

DuckDB es una base de datos analítica embebida, ligera y de alto rendimiento, diseñada para trabajar directamente dentro de tus aplicaciones sin necesidad de un servidor externo. Se utiliza principalmente para análisis de datos, ofreciendo una alternativa rápida y eficiente a herramientas como pandas o sistemas más pesados como PostgreSQL.

## Ventajas
- Tipo de base de datos: Es una base de datos columnar embebida, inspirada en SQLite, pero optimizada para análisis de datos.
- Funcionamiento: Se ejecuta directamente en memoria dentro de tu aplicación (Python, R, C++), sin requerir instalación de un servidor.
- Objetivo: Facilitar consultas analíticas sobre grandes volúmenes de datos de manera rápida y sencilla.


## Principales características
- Embebida: No necesita un servidor, se integra directamente en tu aplicación.
- Columnar: Almacena datos por columnas, lo que mejora el rendimiento en consultas analíticas.
- Compatibilidad: Funciona con Python, R, C++ y Java, integrándose con librerías como pandas.
- SQL estándar: Permite usar consultas SQL tradicionales, lo que facilita la adopción.
- Rendimiento: Optimizada para operaciones analíticas intensivas (joins, agregaciones, filtros).
- Portabilidad: Al ser ligera, se puede usar en laptops, servidores pequeños o entornos cloud sin complicaciones.


## ¿Para qué sirve DuckDB?
- Análisis de datos locales: Ideal para trabajar con datasets grandes en tu computadora sin depender de un servidor externo.
- Procesamiento rápido: Supera las limitaciones de pandas en Python cuando los datos son muy grandes.
- Exploración interactiva: Perfecta para notebooks de Jupyter o entornos de desarrollo donde se requiere experimentar con datos.
- ETL ligero: Puede usarse para transformar y limpiar datos antes de cargarlos en sistemas más grandes.
- Prototipado: Excelente para pruebas rápidas de modelos analíticos sin necesidad de infraestructura compleja.

## Conclusión
DuckDB es una herramienta perfecta para analistas y científicos de datos que necesitan ejecutar consultas rápidas sobre grandes volúmenes de información sin montar un servidor complejo. Si trabajas con Python o R y buscas superar las limitaciones de pandas, DuckDB es una opción muy práctica y eficiente



```python
# ingestion/load_to_duckdb.py
import duckdb
import pandas as pd

# 1. Conectarse a DuckDB (si no existe el archivo, lo crea solo)
con = duckdb.connect("data/nba.duckdb")

# 2. Leer los Excel con pandas (esto ya sabés hacerlo)
df_teams  = pd.read_excel("BBDD_Player.xlsx")
df_players = pd.read_excel("BBDD_Player.xlsx")

# 3. Crear un schema "bronze" para dejar claro que es data sin transformar
con.execute("CREATE SCHEMA IF NOT EXISTS bronze")  # pista: "CREATE SCHEMA IF NOT EXISTS raw"


# 4. Escribir cada DataFrame como tabla en DuckDB
# pista: duckdb puede registrar un dataframe y hacer INSERT directo
# investigá: con.execute("CREATE OR REPLACE TABLE raw.teams AS SELECT * FROM df_teams")

con.execute("CREATE OR REPLACE TABLE bronze.team AS SELECT * FROM df_teams")  # tabla de equipos
con.execute("CREATE OR REPLACE TABLE bronze.player AS SELECT * FROM df_players")  # tabla de jugadores

# 5. Verificar que funcionó
result = con.execute("SELECT COUNT(*) FROM bronze.teams").fetchone()
print(f"Filas en raw.teams: {result[0]}")

con.close()
```
